## Spectrums

Для проведення спектрального аналізу Wave надає функцію `make_spectrum`, яка розраховує частотний склад звуку.

## Гармонійний аналіз звуку літери Н

```python
spectrumN = segmentN.make_spectrum()
```

Спектр можна побудувати за допомогою `plot`

```python
spectrumN.plot()
```

Компонентів з частотою більше 3 kHz досить мало. Можна краще роздивитися графік поставивши верню ланку частоти в 3000.

```python
spectrumN.plot(high=3000)
```

Spectrum надає функцію `peaks`, яка повертає всі обчислені точки-вершини у відсртованому порядку за амплітудою. Отже можна легко визначити значення амплітуд перших гармонік.

```python
peaksN = spectrumN.peaks()
peaksN
```

Отже з отриманих вершин можна встановити, що амплітуда для 1-ої гармоніки становить 18,284 мВ, а 2-ої = 9,582 мВ.
В середовищі MicroCap значення амплітуди 1-ої гармоніки = 17.1, 2-ої = 9,5. Тобто результати майже збігаються.


Визначимо потужність сигналу букви Н через середнє квадратичне.

```python
import numpy as np

amplitN = [item[0] for item in  peaksN]
amplitN = np.array(amplitN)
rmsN = np.sqrt(np.mean(amplitN**2))

print("Потужність за RMS (мкВт) = ", rmsN*100)
```

Тепер визначимо потужність скориставшись теоремою Парселя.

```python
harmN1 = 18.284
harmN2 = 9.582
parselN = harmN1**2 + harmN2**2

print("Потужність за Парселем (мкВт) = ", parselN)
```

Розрахуємо відносну похибку оцінки потужності.

```python
rmsN = rmsN*100
errN = (rmsN - parselN)/rmsN*100

print("Відносна похибка (%) = ", errN)
```

Бачимо, що відносна похибка є набагато меншою за розраховану в MicroCap. Це по-перше, може бути спричинене більшою частотою дискретизації. По-друге, деякий вклад має неточність визначеня амплітуд гармонік з графіків в MicroCap.

```python
amplitA = [item[0] for item in peaksA]
amplitA = np.array(amplitA)
rmsA = np.sqrt(np.mean(amplitA**2))

print("Потужність за RMS (мкВт) = ", rmsA*100)
```

Тепер визначимо потужність скориставшись теоремою Парселя.

```python
harmA1 = 28.543
harmA2 = 6.012
parselA = harmA1**2 + harmA2**2

print("Потужність за Парселем (мкВт) = ", parselA)
```

Розрахуємо відносну похибку оцінки потужності.

```python
rmsA = rmsA*100
errA = (rmsA - parselA)/rmsA*100

print("Відносна похибка (%) = ", errA)
```

Бачимо, що відносна похибка є великою, що зовсім не збігається з результатами в MicroCap.
